```python
import plotly.plotly as py
init_notebook_mode(connected=True)
import cufflinks as cf
import plotly.graph_objs as go
```

```python
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
```

```python
transaction = pd.read_csv("transaction_data.csv")
```

```python
transaction.head()
```

```python
transaction.dtypes
```

```python
transaction['ctot'] = np.cumsum(transaction['tot tonnes'])
```

```python
transaction['DATETIMEIN'] = sorted(transaction['DATETIMEIN'])
```

```python
trace_high = go.Scatter3d(
    x=transaction['DATETIMEIN'],
    y=transaction['ctot'],
    line = dict(color = '#17BECF'),
    opacity = 0.8)

trace_low = go.Scatter(
    x=transaction['DATETIMEIN'],
    y=transaction['ctot'],
    line = dict(color = '#7F7F7F'),
    opacity = 0.8)
data = [trace_high, trace_low]

layout = dict(
    title = "Manually Set Date Range",
    xaxis = dict(
        range = ['2016-12-01','2017-09-28'])
)
```

```python
fig = dict(data = data, layout = layout)
iplot(fig, filename = "Time Series with Rangeslider")
```
